```python
import matplotlib.pyplot as plt

class World:
    def __init__(self, dt):
        self.time = 0.0
        self.dt = dt
        # Zaczynamy z czasem 0.0 w historii
        self.history = [0.0]

    def step(self):
        self.time += self.dt
        self.history.append(self.time)
        

def draw_time(world):
    plt.figure()
    # Tworzymy listę kroków (0, 1, 2...) dla osi X
    steps = list(range(len(world.history)))
    plt.plot(steps,world.history,'o')
    plt.xlabel("Krok symulacji")
    plt.ylabel("Czas świata")
    plt.title("Czas w symulacji (krok po kroku)")
    plt.grid(True)
    plt.show()

world = World(dt = 0.25)

for _ in range(10):
    world.step()
   
print("world.history")
print(world.history)
draw_time(world)
```

```python
Oto rozbicie tego kodu na czynniki pierwsze:
1. Nagłówek i wyrównanie tekstu

print(f"{'KROK':<5} | {'CZAS ŚWIATA':<12} | {'STATUS'}")
:<5 – Oznacza: „Zarezerwuj 5 znaków szerokości i wyrównaj tekst do lewej (<)”.
Dzięki temu słowo "KROK" i liczby pod nim będą w jednej kolumnie.
:<12 – Rezerwuje 12 znaków dla czasu.
Jeśli czas jest krótki (np. 0.5), reszta zostanie wypełniona spacjami.
    
2. Linia oddzielająca
print("-" * 35)
To sprytny trik Pythona: mnożenie tekstu.
Powtarza myślnik 35 razy, tworząc linię pod nagłówkiem.
3. Pętla i formatowanie liczb

Wewnątrz pętli działo się najwięcej magii:
print(f"#{i:02d}  | {world_time:<12.3f} | ✅ OK")
# – To po prostu zwykły znak płotka wypisany przed numerem.{i:02d}:

i – Nasza zmienna z pętli (numer kroku).:
– Rozpoczyna instrukcję formatowania.
    
02 – „Użyj 2 cyfr szerokości, a jeśli liczba jest mniejsza, wolne miejsce wypełnij zerem”.
d – Oznacza liczbę całkowitą (decimal).
Efekt: Zamiast 1, 2, 3... zobaczysz 01, 02, 03.... 
Wygląda to estetycznie w tabeli.
    
{world_time:<12.3f}:world_time – Zmienna przechowująca czas (np. 0.125).
<12 – Wyrównaj do lewej w kolumnie o szerokości 12 znaków..
3f – „Pokaż dokładnie 3 miejsca po przecinku”.
f – Oznacza liczbę zmiennoprzecinkową (float).
    
Dlaczego to jest takie ważne w symulacjach?
W fizyce i programowaniu często operujemy na ułamkach (np. $dt = 0.0001$). 
Bez formatowania .3f Twój ekran zostałby zalany liczbami typu 0.10000000000000002 
(to wynika z tego, jak komputery liczą binarne ułamki). 
Dzięki f-stringom panujesz nad tym, co widzi użytkownik.
    
Podsumowanie symboli w f-stringu:
Wyrównaj do lewej
x:<10
Wyrównaj do prawej 
x:>10
Wyśrodkuj 
x:^10
Wyśrodkuj z gwiazdkami
x:*^20    
.Nf -  N miejsc po przecinku 3.1415:.2f -> 3.140
Nd  - Liczba całkowita z zerami 5:03d  -> 005

Przykład w kodzie
Spójrz na różnicę między trzema typami wyrównania:

Python

tytul = "RAPORT"

print(f"|{tytul:<20}|")  # Do lewej
print(f"|{tytul:>20}|")  # Do prawej
print(f"|{tytul:^20}|")  # WYŚRODKOWANE
Wynik w konsoli:

Plaintext

|RAPORT              |
|              RAPORT|
|       RAPORT       |
3. Dekorowanie (Wypełniacz)
Wyśrodkowanie staje się jeszcze fajniejsze, 
gdy zamiast spacji użyjesz innego znaku (np. gwiazdki lub myślnika), 
aby stworzyć ładny separator w swoich logach:

# Składnia: {zmienna:znak_wypełnienia^szerokość}
print(f"{' KONIEC SYMULACJI ':*^40}")
print(f"{' ETAP 1 ':—^40}")
Wynik:

Plaintext

********** KONIEC SYMULACJI ***********
---------------- ETAP 1 ----------------
```
